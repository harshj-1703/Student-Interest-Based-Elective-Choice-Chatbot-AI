```python
text = "'['spread spectrum cdma', 'communications spread spectrum', 'introduction digital communications', 'sequence spread spectrum', 'spread spectrum performance', 'digital communications spread', 'spread spectrum frequency', 'spectrum cdma', 'spectrum frequency hopping', 'frequency hopping']"
```

```python
words = text.replace("'", '')
words = words.split()
words = [word.strip('.,!;()[]') for word in words]
words = [word.replace("'s", '') for word in words]

#finding unique
unique = []
for word in words:
    if word not in unique:
        unique.append(word)
```

```python
# import required libraries
import numpy as np
from numpy.linalg import norm

array_len = len(user_vectored.toarray())

cosine_similarities = []

# compute cosine similarity
for i in range(0,array_len):
  cosine = np.dot(user_vectored.toarray()[array_len-1],user_vectored.toarray()[i])/(norm(user_vectored.toarray()[array_len-1])*norm(user_vectored.toarray()[i]))
  # print("Cosine Similarity:", cosine)
  cosine_similarities.append(cosine)
```

```python
cosine_similarities
```

```python
from itertools import islice
cosine_similarities = list(islice(cosine_similarities, len(cosine_similarities)-1))
# cosine_similarities
```

```python
index = cosine_similarities.index(max(cosine_similarities))
index
```
